### Task 1: Converting `host_since` from an integer to a datetime object.
In the data dictionary, the datatype of the column is set to be `date`, however it is showing up as an `float64`. So, there must be a mistake in the data dictionary.
The column is defined as: "The date the host/user was created. For hosts that are Airbnb guests this could be the date they registered as a guest."

```python
# Initial data type of the column
airbnb['host_since'].dtype
```

```python
airbnb.host_since.value_counts()
```

Initially, I couldn't figure out how to do this. Any of the values in the `host_since` column (e.g. `39698.0`) are too low to be a credible date (the epoch converter can only go as far as 1/1/1970, which is well before the establishment of AirBnB. Even if you assume that they refer to a time interval, the time interval is too short to be credible - e.g. if you convert `39698.0` in epoch, you get 11 hours. It's unlikely that such a recent member of AirBnB would have a well-reviewed listing. <br> The only believable meaning that I could gather from this is that this column refers to the number of hours that have elapsed since the profile was started - e.g. if you convert `39698.0`, you get 4.5 years. <br> Therefore, I will proceed to convert the values in this column as if they were hours.
<br> I would like this column to reflect the date of the host's initial activity in AirBnB as a date format, so this will be calculated as follows: <br>
`host_since (new values) = date of retrieval of data (DD/MM/YYY) - no. hours elapsed` <br>
The date of retrieval is available from the data dictionary - it is 01/08/2022 or the version 4.3 of the data.



```python
# Save the date when the data was loaded. Save it as a string
date_of_retrieval_str = '01/08/2022'
# Convert this date to the datetime format
date_of_retrieval = pd.to_datetime(date_of_retrieval_str, format = '%d/%m/%Y')
# Create new column
airbnb.host_since_date = 0
# Calculate the corresponding dates based on the hours since the day when the account was created
airbnb.host_since_date = date_of_retrieval - pd.to_timedelta(airbnb.host_since, unit = 'h')

# Convert column to DD/MM/YYYY format
airbnb.host_since_date = airbnb.host_since_date.dt.strftime('%d/%m/%Y')
# Check it worked
airbnb.head()
```

Private rooms
Private rooms are great for when you prefer a little privacy, and still value a local connection. When you book a private room, you'll have your own private room for sleeping and may share some spaces with others. Shared rooms
